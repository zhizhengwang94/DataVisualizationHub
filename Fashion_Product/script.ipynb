# Package Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

# Data Cleaning

```python
data.isna().any()
```

```python
sns.heatmap(data.isna(),cmap='hot',cbar=False)
```

```python
data.nunique()
```

```python
data.replace('Nan',np.nan,inplace=True)
```

# Feature Engineering

```python
data.dropna(axis=0,inplace=True)
```

```python
data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
data["MRP"] = data["MRP"].apply(lambda x: int(x[3:]) if not pd.isna(x) else x)
data['SellPrice'] = data['SellPrice'].astype(int)
data["Sizes"] = data["Sizes"].apply(lambda x: x[5:] if not pd.isna(x) else x)
```

```python
data.dtypes
```

```python
sns.boxplot(x = data["SellPrice"])
```

```python
#For further analysis, we want to remove above outliers
def min_max(dt):
    q1 = dt.quantile(0.25)
    q3 = dt.quantile(0.75)
    iqr = q3 - q1
    mn = q1 - 1.5 * iqr # min value
    mx = q3 + 1.5 * iqr # max value
    return mn, mx
```

```python
mn, mx = min_max(data["SellPrice"])
data = data[(data["SellPrice"] > mn) & (data["SellPrice"] < mx)]
```
