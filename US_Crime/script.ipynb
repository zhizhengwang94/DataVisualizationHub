```python
import numpy as np
import pandas as pd 
```

```python
crimes = pd.read_csv('data/Crimes_With_Dates_Cleaned.csv')
```

```python
crimes.head()
```

```python
crimes_by_year = crimes['Year'].value_counts().sort_index(ascending=True)
crimes_by_month_2020 = crimes[crimes['Year'] == 2020]['Month'].value_counts().sort_index(ascending=True)
crimes_by_month_2021 = crimes[crimes['Year'] == 2021]['Month'].value_counts().sort_index(ascending=True)
crimes_by_month_2022 = crimes[crimes['Year'] == 2022]['Month'].value_counts().sort_index(ascending=True)
crimes_in_January_2021_by_day = crimes[crimes['Year-Month'] == '2021-01']['Day'].value_counts().sort_index(ascending=True)

crimes_by_month_2020
```

```python
import matplotlib.pyplot as plt

plt.plot(crimes_by_year.index, crimes_by_year.values)
plt.title('Crimes committed in the US')
plt.ylabel('Crime count')
plt.xlabel('Time (years)')
plt.show()
plt.clf()

plt.plot(crimes_by_month_2020.index, crimes_by_month_2020.values)
plt.xlabel('Time (months in 2020)')
plt.ylabel('Crime count')
plt.show()
plt.clf()

plt.plot(crimes_by_month_2021.index, crimes_by_month_2021.values)
plt.xlabel('Time (months in 2021)')
plt.ylabel('Crime count')
plt.show()
plt.clf()

plt.plot(crimes_in_January_2021_by_day.index, crimes_in_January_2021_by_day.values)
plt.xlabel('Time (days in January 2021)')
plt.ylabel('Crime count')
plt.show()
plt.clf()

plt.plot(crimes_by_month_2022.index, crimes_by_month_2022.values)
plt.xlabel('Time (months in 2022)')
plt.ylabel('Crime count')
plt.show()
plt.clf()
```
